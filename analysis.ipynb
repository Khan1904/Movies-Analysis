Outliers

```python
Q1 = df['score'].quantile(0.25)
Q3 = df['score'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['score'] < lower) | (df['score'] > upper)]
print('Lower fence:', lower)
print('Upper fence:', upper)
print('Number of outliers:', len(outliers))
```

By using the IQR method we find that the lower fence is at 3.75 whereas the upper fence is 8.95 and there are 51 outliers meaning there are 51 movies that are out of this range. This shows that there are not errors but extreme cases of good and bad movies.

# Charts and Insights

```python
genre_counts = df['genre'].value_counts()
plt.bar(genre_counts.index, genre_counts.values)
plt.title('Genre Distribution')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

This shows the genre distribution chart. the x axis was initially overlapped and i fixed it with this code: plt.xticks(rotation=45) 

**IMDb Score Distribution**

```python
plt.hist(df['score'], bins=20)
plt.title('Score Distribution')
plt.xlabel('Score')
plt.ylabel('Count')
plt.show()
```

**Age Rating Distribution**

```python
rating_counts = df['rating'].value_counts()
plt.bar(rating_counts.index, rating_counts.values)
plt.title('Age Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

**Runtime Distribution**

```python
plt.hist(df['runtime,,'].str.replace(',', '').astype('float'), bins=20)
plt.title('Runtime Distribution')
plt.xlabel('Runtime (minutes)')
plt.ylabel('Count')
plt.show()
```
