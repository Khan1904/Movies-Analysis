Outliers

```python
Q1 = df['score'].quantile(0.25)
Q3 = df['score'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['score'] < lower) | (df['score'] > upper)]
print('Lower fence:', lower)
print('Upper fence:', upper)
print('Number of outliers:', len(outliers))
```

By using the IQR method we find that the lower fence is at 3.75 whereas the upper fence is 8.95 and there are 51 outliers meaning there are 51 movies that are out of this range. This shows that there are not errors but extreme cases of good and bad movies.
